# [1] 재무상태표 전처리

## (a) 유동자산 유동부채 부채총계 자본총계

```python
rawData1 = pd.read_csv('/content/drive/MyDrive/data/SKhynix_data/df1.csv', index_col = 0)
rawData1 = rawData1.drop( ['Unnamed: 1'], axis=1)
rawData1 = rawData1.T
rawData1['유형'] = rawData1['[D210000] Statement of financial position, current/non-current - Consolidated financial statements (Unit: KRW)'].iloc[:,[0]]
rawData1 = rawData1.drop('[D210000] Statement of financial position, current/non-current - Consolidated financial statements (Unit: KRW)', axis=1)
rawData1 = rawData1[ ['유형', '20201231', '20191231', '20181231', '20171231', '20161231', '20151231',
       '20141231', '20131231', '20121231', '20111231', '20101231', '20091231', '20081231'] ]
rawData1 = rawData1.drop(columns = '20091231')
rawData1 = rawData1.drop(columns = '20081231')

rD1 = rawData1[(rawData1['유형'] == '유동자산') | (rawData1['유형'] == '유동부채') | (rawData1['유형'] == '부채총계') | (rawData1['유형'] == '자본총계')]

financial_statement = rD1.rename(columns = {'유형' : '기준 년/월'}) # 유형 컬럼을 기준 년/월로 수정
financial_statement = financial_statement.set_index('기준 년/월')                   # 기준 년/월을 인덱스로 설정
financial_statement = financial_statement.reindex( columns = ['20101231', '20111231', '20121231', '20131231', '20141231', '20151231', '20161231', 
                            '20171231', '20181231', '20191231', '20201231'] ) # 컬럼을 오름차순으로 재정렬
financial_statement.columns = ('2010/12', '2011/12', '2012/12', '2013/12', '2014/12', '2015/12', '2016/12', 
                  '2017/12', '2018/12', '2019/12', '2020/12') # 컬럼명을 년도/월 형식으로 변경
financial_statement = financial_statement.T # Transpose
financial_statement

```

## (b) 매출채권, 재고자산, 매입채무, 유형자산, 무형자산

```python
rD11 = rawData1[(rawData1['유형'] == '매출채권') | (rawData1['유형'] == '재고자산') | (rawData1['유형'] == '매입채무') | (rawData1['유형'] == '유형자산') | (rawData1['유형'] == '무형자산')]

financial_statement2 = rD11.rename(columns = {'유형' : '기준 년/월'}) # 유형 컬럼을 기준 년/월로 수정
financial_statement2 = financial_statement2.set_index('기준 년/월')                   # 기준 년/월을 인덱스로 설정
financial_statement2 = financial_statement2.reindex( columns = ['20101231', '20111231', '20121231', '20131231', '20141231', '20151231', '20161231', 
                            '20171231', '20181231', '20191231', '20201231'] ) # 컬럼을 오름차순으로 재정렬
financial_statement2.columns = ('2010/12', '2011/12', '2012/12', '2013/12', '2014/12', '2015/12', '2016/12', 
                  '2017/12', '2018/12', '2019/12', '2020/12') # 컬럼명을 년도/월 형식으로 변경
financial_statement2 = financial_statement2.T # Transpose
financial_statement2
```

## (c) 재무상태표 폴더에 저장

```python
pd.DataFrame( financial_statement ).to_csv('/content/drive/MyDrive/data/SKhynix_data/1_재무상태표/financial_statement.csv')
pd.DataFrame( rawData1 ).to_csv('/content/drive/MyDrive/data/SKhynix_data/1_재무상태표/rawData1.csv')
pd.DataFrame( financial_statement2 ).to_csv('/content/drive/MyDrive/data/SKhynix_data/1_재무상태표/financial_statement2.csv')
```

# [4] 안정성 지표
(a) 유동비율 & 부채비율

```python
# 유동비율과 부채비율에 필요한 financial_statement 불러오기
financial_statement = pd.read_csv('/content/drive/MyDrive/data/SKhynix_data/1_재무상태표/financial_statement.csv', index_col = 0)
financial_statement
```

```python
current_ratio = []
debt_ratio = []

for i in range(len(financial_statement['유동자산'])):
  current_ratio.append( ((float(financial_statement['유동자산'][i])) / float(financial_statement['유동부채'][i]))*100 )

for j in range(len(financial_statement['부채총계'])):
  debt_ratio.append( ((float(financial_statement['부채총계'][j])) / float(financial_statement['자본총계'][j]))*100 )

dicts = { '유동비율(%)':current_ratio, '부채비율(%)':debt_ratio}
stable_ind = pd.DataFrame(dicts)
stable_ind.set_index(financial_statement.index, inplace=True)

stable_ind.drop(index= stable_ind.index[0] , inplace=True)
pd.DataFrame( stable_ind ).to_csv('/content/drive/MyDrive/data/SKhynix_data/2_안정성 지표/stable_ind.csv')

stable_ind
```

```python
# 안정성지표의 유동비율과 부채비율 저장
pd.DataFrame( stable_ind ).to_csv('/content/drive/MyDrive/data/SKhynix_data/2_안정성 지표/stable_ind.csv')
```

# [5] 수익성 지표


```python
# ROE / ROIC / 영업이익률 계산

ROE = []
ROIC = []
Profit = []

for i in range(11):

  ROE.append( (float(data['당기순이익'][i]) / float(data['자본총계'][i]) *100 ))

for j in range(11):

  ROIC.append( (( float(data['영업이익'][j]) - float(data['법인세비용'][j]) ) / ((float(data['매출채권'][j])+float(data['재고자산'][j])-float(data['매입채무'][j])) +float(data['유형자산'][j])+float(data['무형자산'][j])))*100 )

for k in range(11):

  Profit.append( (float(data['영업이익'][k]) / float(data['매출액'][k]) *100 ))

dicts = { 'ROE(%)': ROE, 'ROIC(%)': ROIC, '영업이익률(%)': Profit}
Profit_ind = pd.DataFrame(dicts)
Profit_ind.set_index(data.index, inplace=True)

Profit_ind.drop(index= Profit_ind.index[0] , inplace=True)
pd.DataFrame( Profit_ind ).to_csv('/content/drive/MyDrive/data/SKhynix_data/2_수익성 지표/Profit_indicator.csv')

Profit_ind
```

```python
# 수익성 지표의 ROE / ROIC / 영업이익률 저장
pd.DataFrame( Profit_ind ).to_csv('/content/drive/MyDrive/data/SKhynix_data/2_수익성 지표/Profit_indicator.csv')
```

# [6] 성장성 지표

```python
# 매출액 증가율 자기자본 증가율 계산
# [당기 - 전기]이기 때문에 보정을 위해 
# 2010년 row에 0을 넣어준다.

inc1 = [0] # 매출액 증가율
inc2 = [0] # 자기자본 증가율

for i in range(10):
  inc1.append( ((float(data['매출액'][i+1]) / float(data['매출액'][i]))*100) - 100 )

for j in range(10):
  inc2.append( ((float(data['자본총계'][j+1]) / float(data['자본총계'][j]))*100) - 100 )

dicts = { '매출액 증가율(%)': inc1, '자기자본 증가율(%)': inc2 }
increase_ind = pd.DataFrame(dicts)
increase_ind.set_index(data.index, inplace=True)

increase_ind.drop(index= increase_ind.index[0] , inplace=True)

increase_ind

pd.DataFrame( increase_ind ).to_csv('/content/drive/MyDrive/data/SKhynix_data/2_성장성 지표/increase_indicator.csv')
```
